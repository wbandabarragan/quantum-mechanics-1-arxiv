# Particle in an infinite potential well

```python
# Importing libraries, # can be used in python to comment code

import numpy as np # np, plt are aliases
import scipy
import matplotlib.pyplot as plt
```

```python
def psi_function(a, n, x):
  
  psi_x = np.sqrt(2/a)*np.sin(n*np.pi*x/a)

  return psi_x
```

```python
x = np.arange(0, 2*np.pi, 0.01)

#print(x)
```

```python
a = 2*np.pi

print(a)
```

```python
y = psi_function(a, 3, x)
```

```python
#print(y)
```

```python
%matplotlib inline

plt.plot(x, y)
```

```python
y1 = psi_function(a, 1, x) # n=1

y2 = psi_function(a, 2, x) # n=2

y3 = psi_function(a, 3, x) # n=3

y4 = psi_function(a, 4, x) # n=4
```

```python
fig, ax = plt.subplots(figsize=(12,10))

ax.plot(x, y1,  label = r'$\psi_1$')
ax.plot(x, y2,  label = r'$\psi_2$')
ax.plot(x, y3,  label = r'$\psi_3$')
ax.plot(x, y4,  label = r'$\psi_4$')

ax.legend(fontsize=18)
```

```python
def psi_function(a, n, x):
    if(x <= 0):
        return 0
    elif(0 <= x <= a):
        return np.sqrt(2/a)*np.sin(n*np.pi*x/a)
    else:
        return 0
```

```python
x_general = np.arange(-2*np.pi, +4*np.pi, 0.01)
```

```python
y_general = []
```

```python
for i in range(len(x_general)):
    y_general.append(psi_function(a, 4, x_general[i]))
```

```python
fig, ax = plt.subplots(figsize=(12,10))

ax.plot(x_general, y_general, c = 'red', label = r'$\psi_1$')
ax.legend(fontsize=18)
```
