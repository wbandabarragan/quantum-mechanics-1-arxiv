4) Now, let's make a simple plot:

```python
# Vectors for the X axis and Y axis

x = np.arange(-2*np.pi, 2*np.pi, 0.01)

y = np.cos(x)

print(x.shape, y.shape)
```

```python
# For the plot we use Matplotlib

plt.figure(figsize = (5,4))

plt.plot(x, y, color = "magenta", linestyle = ":", label = r"$y = \cos(x)$")

plt.xlabel(r"$x$")
plt.ylabel(r"$y$")

plt.legend(loc=1)

plt.show()
```

```python
# Another Y axis

y_1 = x
y_2 = np.sin(x)
```

```python
# For the plot we use Matplotlib

plt.figure(figsize = (6,5))

plt.plot(x, y, color = "red", label = r"$y = \cos(x)$")
plt.plot(x, y_1, color = "blue", label = r"$y = x$")
plt.plot(x, y_2, color = "green", label = r"$y = \sin(x)$")

plt.xlabel(r"$x$")
plt.ylabel(r"$y$")

plt.legend(loc=4)
plt.show()
```

5) Surface plots:


$$z = \sin\left(\sqrt{x^2+y^2}\right)$$

```python
# Let's define a python function

def surface_1(x_2D, y_2D):
  """
  This is a function to create a 2D surface.
  Inputs: x_2D, y_2D
  Outputs: z_2D
  """

  z_2D = np.sin(np.sqrt(x_2D**2 + y_2D**2))

  return z_2D
```

```python
help(surface_1)
```

```python
# We need to create a 2D grid, based on 1D vectors

# Let's create the 1D vectors
x = np.linspace(-10., 10., 400)
y = np.linspace(-10., 10., 400)

# Now we need the 2D grid
x_2D, y_2D = np.meshgrid(x, y)

print(x.shape, x_2D.shape)
print(y.shape, y_2D.shape)

# Call the surface function
z_2D = surface_1(x_2D, y_2D)

print(z_2D.shape)
```

```python
# Plotting

plt.figure()

ax = plt.axes(projection = "3d")

ax.contour3D(x_2D, y_2D, z_2D, 100, cmap = 'magma')

ax.set_xlabel('x')  # Customised labels for axes
ax.set_ylabel('y')
ax.set_zlabel('z')

plt.show()
```

```python
# Plotting

plt.figure(figsize=(5,4))

plt.pcolor(x_2D, y_2D, z_2D, cmap = 'magma')

plt.xlabel('x')  # Customised labels for axes
plt.ylabel('y')

plt.show()
```

```python
# Plotting

plt.figure(figsize=(5,4))

plt.contour(x_2D, y_2D, z_2D, cmap = 'magma')

plt.xlabel('x')  # Customised labels for axes
plt.ylabel('y')

plt.show()
```

```python
# Plotting

plt.figure(figsize=(5,4))

plt.contourf(x_2D, y_2D, z_2D, cmap = 'magma')

plt.xlabel('x')  # Customised labels for axes
plt.ylabel('y')

plt.show()
```
