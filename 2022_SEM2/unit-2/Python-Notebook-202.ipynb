## B) Infinite square well potential:

The potential, $V(x)$, is given by the following piecewise function:

$   
V(x)= 
     \begin{cases}
       0, & 0 \leq x \leq a\\
       \infty, & \rm elsewhere\\
     \end{cases}
$

### Solution:

$\psi_n(x) = \sqrt{\frac{2}{a}} \sin\left(\frac{n\pi}{a}x\right)$

with:

$E_n = \frac{n^2\hbar^2\pi^2}{2ma^2}$


```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.constants as cte
```

```python
# Physical quantities
# https://docs.scipy.org/doc/scipy/reference/constants.html

# Define our SI constants
hbar = cte.hbar

# Let's consider that our particle is an e-
me = cte.m_e

# Width of the potential:
a = 5e-10 # in m = 5A = 0.5nm
```

# Plot potential:

$   
V(x)= 
     \begin{cases}
       0, & 0 \leq x \leq a\\
       \infty, & \rm elsewhere\\
     \end{cases}
$

```python
# Define piece-wise function

def V_function(x, a):

  if(0 <= x <= a):
    return 0
  else:
    return 10000
```

```python
# Vector for x:
x = np.arange(-3e-10, +8e-10, 1.e-11)
#print(x.shape)

# Evaluate the function
# Empty list

y = []

for i in range(len(x)):
  y.append(V_function(x[i], a))

y = np.array(y)
print(x.shape, y.shape)
```

```python
# Plotting

plt.figure(figsize = (6, 5))
plt.plot(x, y)
plt.ylim(0, 1000)
plt.xlabel(r"$x$")
plt.ylabel(r"$V(x)$")
plt.show()
```

### Solution:

$\psi_n(x) = \sqrt{\frac{2}{a}} \sin\left(\frac{n\pi}{a}x\right)$

```python
# Define piece-wise function for psi

def psi_function(x, a, n):

  if(0 <= x <= a):
    psi_n = np.sqrt(2/a)*np.sin(n*np.pi*x/a)
    return psi_n
  else:
    return 0
```

```python
# Call the function for psi_n

# Evaluate the function
# Empty list for n=1
psi_1 = []

for i in range(len(x)):
  psi_1.append(psi_function(x[i], a, 1))

psi_1 = np.array(psi_1)

# Empty list for n=2
psi_2 = []

for i in range(len(x)):
  psi_2.append(psi_function(x[i], a, 2))

psi_2 = np.array(psi_2)

# Empty list for n=3
psi_3 = []

for i in range(len(x)):
  psi_3.append(psi_function(x[i], a, 3))

psi_3 = np.array(psi_3)

# Empty list for n=3
psi_4 = []

for i in range(len(x)):
  psi_4.append(psi_function(x[i], a, 4))

psi_4 = np.array(psi_4)
```

```python
# Plotting

plt.figure(figsize = (6, 5))
plt.plot(x, psi_1)
plt.plot(x, psi_2)
#plt.ylim(0, 1000)
plt.xlabel(r"$x$")
plt.ylabel(r"$\psi(x)$")
plt.show()
```

```python
# Multi-panel figure, ax = axes

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2, figsize = (8,4))

ax1.plot(x, psi_1, linestyle = "-", color = 'red',  label=r"$n= 1$")
ax2.plot(x, psi_2, linestyle = "-", color = 'magenta', label=r"n= 2$")
ax3.plot(x, psi_3, linestyle = "-", color = 'green', label=r"$n = 3$")
ax4.plot(x, psi_4, linestyle = "-", label=r"$n= 4$")

ax1.set_ylim(-8e4, +8e4)
ax2.set_ylim(-8e4, +8e4)
ax3.set_ylim(-8e4, +8e4)
ax4.set_ylim(-8e4, +8e4)

ax1.legend()
ax2.legend()
ax3.legend()
ax4.legend()

plt.show()
```

### Energy ladder:

$E_n = \frac{n^2\hbar^2\pi^2}{2ma^2}$

```python
def En_function(n):

  en = (n**2*hbar**2*np.pi**2)/(2*me*a**2)

  return en

```

```python
# Vector for n

n = np.array([1, 2, 3, 4])
```

```python
# Evaluate energies:

E_n = En_function(n)

print("Energies in [J]:", E_n)
print("Energies in [eV]:", E_n*6.242e18)
```

```python
# Increase in energy between two states

in_En2 = E_n[1]/E_n[0]
in_En3 = E_n[2]/E_n[0]

print(in_En2, in_En3)
```

## Energy ladder E_n vs. n

```python
E_ev = E_n*6.242e18

plt.figure(figsize=(8, 7))

plt.plot(n, E_ev, linestyle =' ', marker = "+")

plt.xlabel(r"$n$")
plt.ylabel(r"$E_n [eV]$")

plt.show()
```

## Add the wiggle factor to the n=1 solution



```python
# Plotting

plt.figure(figsize = (6, 5))
plt.plot(x, psi_1, linestyle = "-", color = 'red',  label=r"$n= 1$")
plt.ylim(-8e4, +8e4)
plt.xlabel(r"$x$")
plt.ylabel(r"$\psi(x)$")
plt.show()
```

### Wiggle factor:

$$\phi(t) =e^{-i\frac{E_n}{\hbar}t}$$

$$\phi(t) =\cos\left(\frac{E_n}{\hbar}t\right) - i\sin\left(\frac{E_n}{\hbar}t\right)$$

```python
# Compute chracteristic frequency

# For n = 1
freq = E_n[0]/hbar

print(freq)
```

```python
# Time vector

t_axis = np.linspace(0, 2*np.pi/freq, 10)
```

```python
print(t_axis)
```

$$\phi(t) =\cos\left(\frac{E_n}{\hbar}t\right)$$

```python
wiggle_factor = np.cos(freq*t_axis)

print(t_axis.shape, wiggle_factor.shape)

print(wiggle_factor)
```

```python
# Plotting the full Psi(x, t)
# Empty list to put all solutions
psi_xt1 = []

for i in range(len(t_axis)):
  psi_xt1.append(psi_1*wiggle_factor[i])

psi_xt1=np.array(psi_xt1)
print(psi_xt1.shape)
```

```python
# Plotting

for j in range(len(t_axis)):
  plt.figure(figsize = (6, 5))
  plt.plot(x, psi_xt1[j,:], linestyle = "-", color = 'red',  label=r"$n= 1$")
  #plt.plot(x, psi_xt1[1,:], linestyle = "-", color = 'red',  label=r"$n= 1$")
  #plt.plot(x, psi_xt1[2,:], linestyle = "-", color = 'red',  label=r"$n= 1$")
  plt.ylim(-8e4, +8e4)
  plt.xlabel(r"$x$")
  plt.ylabel(r"$\psi(x)$")

  #plt.savefig("")
  plt.show()
```
