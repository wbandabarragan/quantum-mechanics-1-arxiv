## **1. Gaussian distributions:**

Consider the Gaussian distribution:

$\rho(x) = A\,e^{−\lambda\,(x−a)^2}$

where: $A = \sqrt{\frac{\lambda}{\pi}}$, $a$, $\lambda$ are positive real constants.

Sketch $\rho(x)$ versus $x$ for different values of $A$, $a$ and $\lambda$.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define a function for a 1D Gaussian

def rho_x(l, a, x):
  """
  This is a function to compute Gaussian PDFs.
  Inputs: l -> lambda, a -> x displacement, x -> vector
  """
  pdf_x = np.sqrt(l/np.pi)*np.exp(-l*(x - a)**2)

  return pdf_x
```

```python
#help(rho_x)

# Define an x vector
x = np.linspace(-10., 10., 1000)

#print(x.shape)

# Define the constants
l_1 = 2.
a_1 = 0.5

l_2 = 4.
a_2 = 3.5

l_3 = 0.5
a_3 = -2.5

# Evaluate the function using x
y_1 = rho_x(l_1, a_1, x)
y_2 = rho_x(l_2, a_2, x)
y_3 = rho_x(l_3, a_3, x)
```

```python
# Plotting

plt.figure(figsize=(5,4))

plt.plot(x, y_1, color = "red")
plt.plot(x, y_2, color = "blue")
plt.plot(x, y_3, color = "magenta")

plt.xlabel(r"$x$")
plt.ylabel(r"$\rho(x)$")

plt.show()
```

## **2. Piece-wise Wavefunctions:**

Sketch the following $\Psi(x)$ at $t=0$ (Problem 1.4 in Griffiths):

$$\Psi(x,0)=\left\{
\begin{array}{ll}
      A(x/a) & 0 \leq x \leq a \\
      A(b-x)/(b-a) & a \leq x \leq b \\
      0 & \rm otherwise \\
\end{array}
\right. $$




```python
# Define piece-wise function

def wave_piecewise(x, a, b, A):
  """
  Function that returns piece-wise values for psi(x)
  Inputs: x -> vector, a, b, A -> parameters
  Output: y -> vector
  """
  # Use conditionals
  if(0. <= x <= a):
    return A*(x / a)
  elif(a <= x <= b):
    return A*(b - x)/(b - a)
  else:
    return 0
```

```python
# Define a vector for x:
x = np.arange(-3., 6., 0.01)

# Input parameters
a_1 = 1.
b_1 = 3.
A_1 = 3.

a_2 = 2.
b_2 = 3.
A_2 = 5.

# Evaluate the function
#y = wave_piecewise(x, a_1, b_1, A_1) # This prints an error
```

```python
# Evaluate the function
# Empty list
y = []
y_2 = []

# Evaluate using a loop
for j in range(len(x)):
  #print(j)
  y.append(wave_piecewise(x[j], a_1, b_1, A_1))
  y_2.append(wave_piecewise(x[j], a_2, b_2, A_2))

# Convert to an array
y = np.array(y)
y_2 = np.array(y_2)
```

```python
# Plotting

plt.figure(figsize=(5,4))

plt.plot(x, y, color = "brown")
plt.plot(x, y_2, color = "pink")

plt.xlabel(r"$x$")
plt.ylabel(r"$\psi(x)$")

plt.show()

```

```python
# Plotting

plt.figure(figsize=(5,4))

plt.plot(x, y**2, color = "brown")
plt.plot(x, y_2**2, color = "pink")

plt.xlabel(r"$x$")
plt.ylabel(r"$|\psi(x)|^2$")

plt.show()
```

## **3. Time-dependent Wavefunction:**

Consider the wave function (Problem 1.5 in Griffiths):

$\Psi(x,t) = A\,e^{-\lambda|\,x|}\,e^{-i\omega t}$

where $A=\sqrt{\lambda}$, $\lambda$, and $\omega$ are positive real constants.

Sketch $|\Psi|^2$ versus $x$ for $\lambda$ different values, and different times.

```python
# Define the function

def psi_xt(l, w, x, t):
  """
  This is a funtion to get an abs valued wave.
  Inputs: l -> lambda, w -> omega, x -> spatial vector, t -> time vector
  Output: y_psi
  """

  y_psi = np.sqrt(l)*np.exp(-l*abs(x))*np.cos(w*t)

  return y_psi
```

```python
# Define an x vector
x = np.arange(-10., 10., 0.01)

# Fix the parameters

l_1 = 1.
w_1 = 1.
t = 0.
t_2 = 0.9

# Evaluate the function
y = psi_xt(l_1, w_1, x, t)
y_2 = psi_xt(l_1, w_1, x, t_2)
```

```python
# Plotting

plt.figure(figsize=(5,4))

plt.plot(x, y, color = "brown", label = r"$t=0$")
plt.plot(x, y_2, color = "blue", label = r"$t=0.9$")

plt.xlabel(r"$x$")
plt.ylabel(r"$\psi(x)$")

plt.legend()

plt.show()
```

```python
# Plotting

plt.figure(figsize=(5,4))

plt.plot(x, y**2, color = "brown", label = r"$t=0$")
plt.plot(x, y_2**2, color = "blue", label = r"$t=0.9$")

plt.xlabel(r"$x$")
plt.ylabel(r"$|\psi(x)|^2$")

plt.legend()

plt.show()
```


$$\psi(x) = A\,e^{-\frac{x^2}{2\sigma^2}}\sin(kx)$$

```python
# Define a wave packet function
def wave_packet(x, A, s, k):
  """
  Function to create a wave packet.
  Inputs: A, k, s -> parameters, x -> vector
  Output: psi_x -> wave packet
  """
  psi_x = A*np.exp(-x**2/(2*s**2))*np.sin(k*x)

  return psi_x
```

```python
# Vector for x
x = np.linspace(-15., 15., 1000)

# Define the parameters:
A = 1.
s = 2.
k = 3.*np.pi

# Get the y vector
y = wave_packet(x, A, s, k)
```

```python
# Plotting

plt.figure(figsize=(5,4))

plt.plot(x, y, color = "blue", label = r"$t=0$")

plt.xlabel(r"$x$")
plt.ylabel(r"$\psi(x)$")

plt.legend()

plt.show()
```

```python
# Plotting

plt.figure(figsize=(5,4))

plt.plot(x, y**2, color = "blue", linestyle = "-", label = r"$t=0$")

plt.xlabel(r"$x$")
plt.ylabel(r"$|\psi(x)|^2$")

plt.legend()

plt.xlim(-5., +5.)
plt.grid()

plt.show()
```

### Moving wave packet:

$$\psi(x,t) = A\,e^{-\frac{(x - t)^2}{2\sigma^2}}\sin(kx - \omega t)$$

```python
# Define a wave packet function
def wave_packet_t(x, t, A, s, k, w):
  """
  Function to create a wave packet.
  Inputs: A, k, s -> parameters, x -> vector
  Output: psi_x -> wave packet
  """
  psi_xt = A*np.exp(-(x - t)**2/(2*s**2))*np.sin(k*x - w*t)

  return psi_xt
```

```python
# Define the new parameters
w = 3*np.pi

# Time vector
t = np.linspace(0., 10., 50)

# Evaluate the packet:
y0 = wave_packet_t(x, t[2], A, s, k, w)
```

```python

# Plotting

plt.figure(figsize=(5,4))

plt.plot(x, y, color = "blue", label = r"$t=0$")
plt.plot(x, y0, color = "cyan", label = r"$t>0$")

plt.xlabel(r"$x$")
plt.ylabel(r"$\psi(x)$")

plt.legend()

plt.show()
```

```python
# For loop for time

for j in range(len(t)):

  # Figure environment
  plt.figure(figsize = (8,6))

  plt.plot(x, wave_packet_t(x, t[j], A, s, k, w), color = "magenta", linestyle = "-", linewidth = 2)

  plt.xlim(-15., +15.)
  plt.ylim(-1.2, +1.2)

  plt.xlabel(r"$x$")
  plt.ylabel(r"$\psi(x,t)$")

  plt.grid()

  plt.savefig("/content/drive/MyDrive/Colab Notebooks/wave_packets/wp_{:02d}.png".format(j))

  plt.close()
```

## **5. 2D Gaussian function**

$$f(x,y) = \frac{1}{2\pi\sigma_x\sigma_y} \exp\left(-\left(\frac{(x - x_0)^2}{2\sigma_X^2} + \frac{(y - y_0)^2}{2\sigma_Y^2} \right)\right).$$

```python
# Define a function

def gaussian_2d(x_2d, y_2d):
  """
  This is a function to evaluate a 2D Gaussian.
  Inputs: x_2d, y_2d -> coordinate matrices from the mesh
  Output: g_2d -> 2D Gaussian
  """

  g_2d = (1/(2*np.pi*sigma_x*sigma_y) * np.exp(-(x_2d**2/(2*sigma_x**2) + y_2d**2/(2*sigma_y**2))))

  return g_2d
```

```python
# Define the input parameters
sigma_x = sigma_y = 2.

# 1D vectors for x and y
x = np.linspace(-10., 10., 1000)
y = np.linspace(-10., 10., 1000)

# 2D mesh grid
x_2d, y_2d = np.meshgrid(x, y)

# Evaluate the Gaussian

z_2d = gaussian_2d(x_2d, y_2d)

print(x_2d.shape, y_2d.shape, z_2d.shape)
```

```python
# 3D plotting

plt.figure()

ax = plt.axes(projection ="3d")

ax.plot_surface(x_2d, y_2d, z_2d, cmap = "magma")

plt.show()
```

```python
# Define a function

def gaussian_2d(x_2d, y_2d):
  """
  This is a function to evaluate a 2D Gaussian.
  Inputs: x_2d, y_2d -> coordinate matrices from the mesh
  Output: g_2d -> 2D Gaussian
  """

  g_2d = (1/(2*np.pi*sigma_x*sigma_y) * np.exp(-(x_2d**2/(2*sigma_x**2) + y_2d**2/(2*sigma_y**2))))*np.cos(2*np.pi*x_2d)*np.cos(2*np.pi*y_2d)

  return g_2d
```

```python
# Evaluate the Gaussian

z_2d = gaussian_2d(x_2d, y_2d)

print(x_2d.shape, y_2d.shape, z_2d.shape)
```

```python
# 3D plotting

plt.figure(figsize=(8,6))

ax = plt.axes(projection ="3d")

ax.plot_surface(x_2d, y_2d, z_2d, cmap = "magma")

ax.view_init(60, 30)

plt.show()
```
