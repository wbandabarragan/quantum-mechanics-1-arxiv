## **9. Problem (wave function):**

Consider the wave function:

$\Psi(x,t) = A\,e^{-\lambda|\,x|}\,e^{-i\omega t}$

where $A=\sqrt{\lambda}$, $\lambda$, and $\omega$ are positive real constants.

Sketch $|\Psi|^2$ versus $x$ for different values.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# wave function at t=0

def psi_x(l, x):
    
    # l is lambda
    A = np.sqrt(l)
    
    # psi function
    psi = (A**2)*np.exp(-l*np.abs(x))
    
    return psi
```

```python
# Plotting

x_axis = np.arange(-10, 10, 0.01)

l_1 = 2
l_2 = 1

# The y values.
y_axis1 = psi_x(l_1, x_axis)
y_axis2 = psi_x(l_2, x_axis)
```

```python
plt.figure(figsize=(8,6))

plt.plot(x_axis, y_axis1, label=f'$\lambda$ ={l_1}')
plt.plot(x_axis, y_axis2, label=f'$\lambda$ ={l_2}')

plt.legend()
plt.show()
```

```python
# wave function at any t

def psi_xt(l, x, t):
    
    # l is lambda
    A = np.sqrt(l)
    
    # psi function plus the wiggle factor
    psi = (A**2)*np.exp(-l*np.abs(x))*np.cos(t)
    
    return psi
```

```python
# The y values.
t_0 = 0
t_1 = 1

y_t0 = psi_xt(l_1, x_axis, t_0)
y_t1 = psi_xt(l_2, x_axis, t_1)
```

```python
plt.figure(figsize=(8,6))

plt.plot(x_axis, y_t0, label=f'$t$ ={t_0}')
plt.plot(x_axis, y_t1, label=f'$t$ ={t_1}')

plt.legend()
plt.show()
```

## **10. Problem: 2D wavefunction**


Gaussian wavefunction:

A particular example of a two-dimensional Gaussian function is
$$f(x,y) = A \exp\left(-\left(\frac{(x - x_0)^2}{2\sigma_X^2} + \frac{(y - y_0)^2}{2\sigma_Y^2} \right)\right).$$

```python
# Define Gaussian function

sigma_x = 2.
sigma_y = 2.

def gf(x, y):
    
    return (1/(2*np.pi*sigma_x*sigma_y) * np.exp(-(x**2/(2*sigma_x**2) + y**2/(2*sigma_y**2)))) # Must be indented
```

```python
# Create a grid:

x = np.linspace(-10, 10, 30)
y = np.linspace(-10, 10, 30)

X, Y = np.meshgrid(x, y)

print(x.shape, y.shape, X.shape, Y.shape)

# surface
Z = gf(X, Y)
```

```python
# Plotting 

fig = plt.figure(figsize=(8,6))

ax = plt.axes(projection='3d')

ax.contour3D(X, Y, Z, 100, cmap='magma')

plt.legend()
plt.show()
```

```python
# Plotting 

fig = plt.figure(figsize=(8,6))

plt.contourf(X, Y, Z, 100, cmap ='magma')

plt.colorbar()
plt.show()
```

## **11. Problem: Gaussian wavepacket**

```python
def fwave(x):  # Which we can call later on
    return np.exp(-x**2/4)*np.sin(2*np.pi*4*x) # Must be indented
```

```python
# x-axis
x = np.linspace(-5, 5, 1000)
y = fwave(x)
```

```python
#plotting
plt.figure(figsize=(8,6))

plt.plot(x, y, label="Gaussian wp.")

plt.legend()
plt.show()
```
