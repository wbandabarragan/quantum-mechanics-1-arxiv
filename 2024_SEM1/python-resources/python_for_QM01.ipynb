## 3. Vector generation:


- np.arange()

- np.linspace()

```python
# Vectors for the X axis and Y axis

x = np.arange(-2*np.pi, 2*np.pi, 0.01)

k1 = 1. # wavenumber -> spatial freq.
k2 = 3. # wavenumber -> spatial freq.

y1 = np.cos(k1*x)
y2 = np.cos(k2*x)

print(x.shape, y1.shape, y2.shape)
```

## 3. 1D plots:

```python
# For the plot we use Matplotlib

plt.figure(figsize = (6,5))

plt.plot(x, y1, color = "magenta", linestyle = ":", label = r"$y = \cos(x)$")
plt.plot(x, y2, color = "blue", linestyle = "-", label = r"$y = \cos(10x)$")

plt.xlabel(r"$x$")
plt.ylabel(r"$y$")

plt.xlim(-2*np.pi, +4*np.pi)

plt.legend(loc=4)

plt.show()
```

## 4. 2D Meshgrids:

```python
# We need to create a 2D grid, based on 1D vectors

# Let's create the 1D vectors
x = np.linspace(-10., 10., 1000)
y = np.linspace(-10., 10., 1000)

#print(x)

# Now we need the 2D grid
x_2d, y_2d = np.meshgrid(x, y)

##print(x.shape, x_2d.shape)
print(y_2d.shape)
```

## 4. 2D plots:

Surface plots:

$$z_1(x,y) = \sin(x) + \cos(y)$$

$$z_2(x,y) = \sin\left(\sqrt{x^2+y^2}\right)$$

```python
# Let's define a python function

def surface_1(x_2d, y_2d):
  """
  This is a function to create a 2D surface.
  Inputs: x_2d, y_2d
  Outputs: z_2d
  """

  z_2d = np.sin(x_2d) + np.cos(y_2d)

  return z_2d
```

```python
help(surface_1)
```

```python
# Call the surface function

z_2d = surface_1(x_2d, y_2d)

print(z_2d.shape)
```

```python
# We need to add this line for 3D plots

from mpl_toolkits.mplot3d import Axes3D
```

```python
# Plotting

fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x_2d, y_2d, z_2d, cmap = "viridis")

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
# Plotting a projection of the surface

plt.figure(figsize=(5,4))

plt.pcolor(x_2d, y_2d, z_2d, cmap = "viridis")

plt.show()
```

$$z_2(x,y) = \sin\left(\sqrt{x^2+y^2}\right)$$

```python
# Let's define a python function

def surface_2(x_2d, y_2d):
  """
  This is a function to create a 2D surface.
  Inputs: x_2d, y_2d
  Outputs: z_2d
  """

  z_2d = np.sin(np.sqrt(x_2d**2 + y_2d**2))

  return z_2d
```

```python
# Call the surface function

z_2d_example2 = surface_2(x_2d, y_2d)

print(z_2d_example2.shape)
```

```python
# Plotting

fig = plt.figure(figsize=(6,5))
ax = fig.add_subplot(111, projection='3d')

ax.plot_surface(x_2d, y_2d, z_2d_example2, cmap = "inferno")

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

```python
# Plotting a projection of the surface

plt.figure(figsize=(5,4))

plt.pcolor(x_2d, y_2d, z_2d_example2, cmap = "inferno")

plt.show()
```
