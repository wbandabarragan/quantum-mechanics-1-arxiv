```python
import numpy as np
import matplotlib.pyplot as plt
```

## Define axes:

```python
x = np.arange(-np.pi, np.pi, 0.001)

print(x.shape)
```

## Define piece-wise function:

```python
def function(x):
    
    if(-np.pi <= x <= -0.5*np.pi):
        return 0
    
    elif(-0.5*np.pi <= x <= +0.5*np.pi):
        return 1
    
    else:
        return 0
```

```python
y = [] #empty list

for i in range(len(x)):
    y.append(function(x[i]))

y = np.array(y) # list to array
```

```python
print(x.shape, y.shape)
```

## Figure:


```python
plt.figure(figsize=(8,4))
plt.plot(x, y)
plt.show()
```

```python
def function_Fourier(x, n):
    f = np.zeros(x.size)
    f = f + 0.5
    
    for i in range(1, n+1):
        y = (2/np.pi)*((-1)**(i-1))*(1/(2*i-1))*np.cos((2*i -1 )*x)
        f += y
        
    return f
```

## Fourier plot:

```python
y_fourier1 = function_Fourier(x, 1)
y_fourier5 = function_Fourier(x, 5)
y_fourier10 = function_Fourier(x, 10)
```

```python
plt.figure(figsize=(8,4))
plt.plot(x, y)
plt.plot(x, y_fourier1)
plt.plot(x, y_fourier5)
plt.plot(x, y_fourier10)
plt.show()
```
